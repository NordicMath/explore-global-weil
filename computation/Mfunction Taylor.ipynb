```python
def taylorsum(z, s, anlist):
    return sum(an * (z - s)^n for n, an in enumerate(anlist))
```

```python
def taylorsumd(z, s, d, anlist):
    return sum(factorial(n + d)/factorial(n) * an * (z - s)^n for n, an in enumerate(anlist[d:]))
```

```python
def taylorsumdf(z, s, d, anlist):
    return sum(binomial(n + d, n) * an * (z - s)^n for n, an in enumerate(anlist[d:]))
```
